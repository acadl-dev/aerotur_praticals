```python
from dash import Dash, html, dcc, Input, Output, State
import plotly.express as px
import pandas as pd

# carregar os dados
df = pd.read_csv("Amprion.csv")


# criar o app
app = Dash(__name__)

# anos disponiveis
anos_disponiveis = sorted(df['Date'].unique())


# layout da aplicação
app.layout = html.Div(children=[
    html.H1(children='Título'),
    html.Div(children='Texto de apresentação.'),
    dcc.Dropdown(
        id='ano-dropdown',
        options=[{'label': i, 'value': i} for i in anos_disponiveis],
        value=anos_disponiveis[-1]
    ),
    dcc.Graph(id='grafico')
])

# Callback para atualizar o gráfico baseado na seleção do dropdown
@app.callback(
    Output('grafico', 'figure'),
    [Input('ano-dropdown', 'value')]
)

def update_output(value):
    # Filtrar os dados
    dff = df[df['Column1'] == value]

    # Criar o gráfico
    fig = px.bar(dff, x='Column2', y='Column3', title='Título do gráfico')

    return fig
    
# Retornar a figura
if __name__ == '__main__':
    app.run(debug=True)
```
